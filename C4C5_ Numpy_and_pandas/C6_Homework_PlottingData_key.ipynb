```python
# Import the libraries you'll need
import pandas as pd
import numpy as np
```

```python
# Import the data and metadata we used in classes 6 and 7
# from the files 'GSE88741-expression.csv' and 'GSE88741-metadata.xlsx'
# Make sure you set the columns gene and sample names as the index
df = pd.read_csv('data/GSE88741-expression.csv', index_col=0)
meta = pd.read_excel("data/GSE88741-metadata.xlsx", index_col=1)
```

```python
# We haven't used the df.sample() method yet. Read the help file and
# sample 200 random rows (genes). Set the random state to 333.
df_sample = df.sample(200, axis = 0, random_state = 333)
```

```python
# Get a list of the sample names
lines = meta.index

# Trim the last two characters off the end of each sample name ('_1', etc.)
# to leave just the name of the cell line
lines = lines.str[:-2]

# Make a new column of data in the meta table with the cell lines
meta['CellLine'] = lines
```

```python
# Get the index names from the meta table
columns = meta.index

# set the column names of df_sample to 'columns'
df_sample.columns = columns
```

The next cell removes low count genes like we did in Class 6, but using a different threshold and a different method. We then use the df.transform() method to convert all of the read counts to log base 2, and then transpose the data. Finally we concatenate the data to the metadata and save the resulting file. 

Make sure you understand the steps below.

```python
## Adjusting data to deal with low counts and zero read values

# Calculate the median of each gene for df_sample
dfmed = df_sample.median(1)
# Make a new df from df_sample of genes with a median read count of more than 5
# Note we need to make a copy, otherwise we will only have a "view" of the data
dfnz = df_sample.loc[dfmed > 5,].copy()
# Make a mask of all of the locations of zero counts
mask1 = (dfnz == 0)

# Nudge the values that have 0 counts to be 0.01.
dfnz[mask1] = 0.01
# Do the log2 transform
df_log2 = dfnz.transform(np.log2)

dfT = df_log2.T
dfT = pd.concat([meta,dfT],axis=1)
```

```python
dfT
```

```python
# Group dfT by CellLine and calculate the mean of each gene for each line
df_group = dfT.groupby('CellLine').mean()
# Transpose the data so the samples are in the columns again
df_group = df_group.T
```

```python
# Get the three genes with the lowest ANOVA scores from the final Hands On section in Class 6
genes_of_interest = ['PMEL','TYRP1','SERPINE2']
```

```python
df_log2.to_csv('df_log2.csv')
dfT.to_csv('df_T.csv')
df_group.to_csv('df_group.csv')
with open("genes.txt", "w") as output:
    output.write(str(genes_of_interest))
```
